```python
import numpy as np

# Buat data jumlah kamar
bedrooms = np.array([1,1,2,2,3,4,4,5,5,5])

# Data harga rumah
house_price = np.array([15000, 18000, 27000, 34000, 50000, 68000, 65000, 81000,85000, 90000])
```

```python
# Menampilkan data dalam bentuk scatter plot/visualisasi
import matplotlib.pyplot as plt

# Untuk memerintah plot agar dihasilkan di notebook bukan popup terpisah
%matplotlib inline

plt.scatter(bedrooms,house_price)
```

```python
from sklearn.linear_model import LinearRegression

# Ubah bentuk array jadi matriks satu colom
bedrooms = bedrooms.reshape(-1,1)
# Panggil fungsi dgn objek
# Latih dengan Regression.fit ()
linreg = LinearRegression()
linreg.fit(bedrooms,house_price)
```

```python
import matplotlib.pyplot as plt

# Menampilkan scatter plot
plt.scatter(bedrooms, house_price)

# Menampilkan garis regresi linear
plt.plot(bedrooms, linreg.predict(bedrooms), color='red')

# Memberikan label sumbu x dan y
plt.xlabel('Jumlah Kamar')
plt.ylabel('Harga Rumah')

# Menampilkan plot
plt.show()
```
